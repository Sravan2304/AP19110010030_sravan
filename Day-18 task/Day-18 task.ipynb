```python
import pandas as pd
```

```python
df = pd.read_csv("food.csv")
df
```

```python
import matplotlib.pyplot as plt
```

```python
# Histogram
plt.hist(df["Food"], width = 0.5)
plt.xlabel("Type of food")
plt.ylabel("Frequency of food")
plt.show()
```

```python
# Bar Graph
plt.bar(df["Juice"],df["Age"],color = "c", width = 0.2)
plt.xlabel("Type of Juice")
plt.ylabel("Age")
plt.show()
```

```python
# Pie Chart 
plt.pie(df["Age"][:10],labels = df["Gender"][:10], autopct = "% 1.1f%%")
plt.show()
```

```python
# Box plot
plt.boxplot(df["Age"])
plt.show()
```

```python
# Line Plot
plt.plot(df["Age"][:10])
plt.xlabel("Frequency")
plt.ylabel("Age")
plt.show()
```

```python
# Scatter Plot
plt.scatter(df["Age"][:20],df["Food"][:20])
plt.xlabel("Age")
plt.ylabel("Type of food")
plt.show()
```

```python
# Sub Plots
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2, figsize = (9,5))

sns.kdeplot(df['Age'], ax = ax1)
sns.kdeplot(df['Age'], ax = ax2)

```
